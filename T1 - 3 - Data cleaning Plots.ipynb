# Plots y visualizacion de los datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../resources/datasets/customer-churn-model/Customer Churn Model.txt')
```

```python
data.head()
```

```python
% matplotlib inline
```

## Scatter Plot

```python
# Nube de puntos
data.plot(kind='scatter', x='Day Mins', y='Day Charge')
```

```python
data.plot(kind='scatter', x='Night Mins', y='Night Charge')
```

```python
figure, axs = plt.subplots(2,2, sharey=True, sharex=True)
data.plot(kind='scatter', x='Day Mins', y='Day Charge', ax=axs[0][0])
data.plot(kind='scatter', x='Night Mins', y='Night Charge', ax=axs[0][1])
data.plot(kind='scatter', x='Day Calls', y='Day Charge', ax=axs[1][0])
data.plot(kind='scatter', x='Night Calls', y='Night Charge', ax=axs[1][1])
```

## Histogramas de frecuencias

```python
"""
    k => Regla de Sturges para el bins
    ceil() => Funcion de Numpy para redondear numero decimal hacia arriba
    3333 => Numero de filas de mi dataset
"""

k = int(np.ceil(1+np.log2(3333)))
plt.hist(data['Day Calls'], bins=k)
plt.title('Histograma de llamadas al dia')
plt.xlabel('N. llamadas al dia')
plt.ylabel('Frecuencia')
```



## Boxplt, diagrama de caja y bigotes

```python
plt.boxplot(data['Day Calls'])
plt.ylabel('Numero de llamadas diarias')
plt.title('Boxplot de las llamadas diarias')
```
