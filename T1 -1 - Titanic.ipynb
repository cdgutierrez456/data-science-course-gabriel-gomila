# Carga de datos a travez de la funcion open

Nota: Para hacer carga de archivos muy grandes, pandas no es la libreria ideal ya que esta libreria hacer una carga de golpe de todo el archivo, por el contrario, la funcion open permite leer el contenido linea a linea para segmentarlo en bloques o segun como se requiera.

```python
filename_churn = 'customer-churn-model/Customer Churn Model.txt'
```

```python
url_file = os.path.join(relative_mainpath, filename_churn)
data3 = open(url_file, 'r')
```

```python
"""
    readline() => Lee linea a linea
    strip() => Se usa para remover los espacios en blanco tanto al principio como al final de la linea
    split() => Divide en pequenios fragmentos por un delimitador
"""
cols = data3.readline().strip().split(',')
```

```python
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
main_dict
```

```python
for line in data3:
    values = line.strip().split(',')
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1
print("El data set tiene %d files y %d columnas" % (counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

Es importante saber que en este caso hemos usado el modo "r" de de la funcion open(), sin embargo, es justo tener cuidado para el caso en que quiera abir en modo "w" ya que eso hace que el contenido previo del archivo sea eliminado y reemplazado por el nuevo contenido. Es mejor, si queremos modificar el documento, abrirlo en modo "a" (append) para hacer la insercion de la nueva informacion, al final del archivo

## Lectura y escritura de ficheros

```python
infile = url_file
outfile = os.path.join(relative_mainpath, 'customer-churn-model/Tab Customer Churn Model.txt')
```

```python
with open(infile, 'r') as infile1:
    with open(outfile, 'w') as outfile1:
        for line in infile1:
            fields = line.strip().split(',')
            outfile1.write("\t".join(fields))
            outfile1.write('\n')
```

```python
df4 = pd.read_csv(outfile, sep='\t')
df4
```
