# Variables dummy

Las variables dummy son nuevas columnas que se agregan al set de datos que consiste en la separacion de columnas categoricas. Ej: columna 'sex', las categorias son 'male' y 'female', al momento de hacer dummy de esta columna, se crean 2 nuevas columnas. 
Las columnas se crean de acuerdo a la cantidad de categorias existentes.

```python
dummy_sex = pd.get_dummies(data['sex'], prefix='sex')
```

```python
dummy_sex
```

```python
column_name=data.columns.values.tolist()
column_name
```

```python
"""
    Con el axis en 1, especificamos que se elimine la columna
"""
data = data.drop(['sex'], axis=1)
```

```python
data = pd.concat([data, dummy_sex], axis=1)
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis=1)
    df = pd.concat([df, dummy], axis=1)
    return df
```
